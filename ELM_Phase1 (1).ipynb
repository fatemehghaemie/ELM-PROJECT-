**ماشین یادگیری شدید (ELM)** نوعی **شبکه‌ی عصبی پیش‌خور با یک لایه‌ی پنهان (SLFN)** است که در آن وزن‌ها و بایاس‌های لایه‌ی پنهان به‌صورت **تصادفی مقداردهی و ثابت نگه داشته می‌شوند** و فرآیند آموزش تنها شامل **محاسبه‌ی تحلیلی وزن‌های خروجی** با استفاده از **شبه وارون مور-پنروز (Moore–Penrose Inverse)** است؛ بدین ترتیب، برخلاف شبکه‌های عصبی متداول، **نیازی به به‌روزرسانی تکراری پارامترها از طریق گرادیان و الگوریتم پس‌انتشار** وجود ندارد.

##  مراحل الگوریتم


1. **آماده‌سازی داده‌ها:**

   مجموعه ورودی‌ها به‌صورت ماتریس‌های  
   $X \in \mathbb{R}^{N \times d}$ و $Y \in \mathbb{R}^{N \times m}$  
    که در آن:
   - $N$: تعداد نمونه‌ها  
   - $d$: تعداد ویژگی‌ها  
   - $m$: تعداد کلاس‌ها یا ابعاد خروجی  

   در مسائل دسته‌بندی، $Y$ معمولاً به‌صورت بردار **وان‌هات (One–Hot)** باید نمایش داده شود.

---

2. **مقداردهی تصادفی پارامترهای لایه‌ی پنهان:**

   وزن‌ها و بایاس‌ها به‌طور تصادفی انتخاب شده و در فرایند آموزش ثابت می‌مانند.

   $$
   W \in \mathbb{R}^{d \times L}, \quad b \in \mathbb{R}^{1 \times L}
   $$


---

3. **محاسبه‌ی ماتریس خروجی لایه‌ی پنهان:**

   با اعمال تابع فعال‌ساز $g(\cdot)$ داریم:

   $$
   H = g(XW + b)
   $$

   که در آن $H \in \mathbb{R}^{N \times L}$ ماتریس ویژگی‌های پنهان است.
<font color="blue" size = 4>
سوال (۱) :
در الگوریتم ELM، بایاس لایه‌ی پنهان به صورت
$
b \in \mathbb{R}^{1 \times L}
$
تعریف شده است.
و هنگام محاسبه ماتریس خروجی لایه‌ی پنهان داریم
$
H = g(X W + b)
$. <br>
اما ابعاد $b$ با $X W \in \mathbb{R}^{N \times L}$ همخوانی ندارد.  
 در کامپیوتر و هنگام استفاده از کتابخانه‌هایی مثل NumPy، جمع $XW + b$ چگونه بدون خطا انجام می‌شود؟  <br>
راهنمایی : برای پاسخ به سوال بالا در رابطه با Broadcasting مطالعه کنید و آن را به طور دقیق در گزارش خود شرح دهید.
</font>

---

4. **محاسبه‌ی وزن‌های خروجی $\beta$:**

   - **بدون منظم‌سازی:**
     $$
     \beta = H^{+} Y
     $$
   - **با منظم‌سازی ریج (Ridge):**
     $$
     \beta = (H^{T}H + \lambda I)^{-1} H^{T}Y
     $$
   
   در رابطه‌ی دوم، $\lambda > 0$ پارامتر منظم‌سازی است.

---

5. **پیش‌بینی خروجی:**

   برای نمونه‌های آزمایشی $X_{\text{test}}$ داریم:

   $$
   \hat{Y} = H_{\text{test}} \, \beta
   $$

   و در مسائل دسته‌بندی برچسب نهایی با بیشترین مقدار خروجی تعیین می‌شود:

   $$
   \text{class} = \arg\max (\hat{Y})
   $$

- `sigmoid`:  $\sigma(z) = \frac{1}{1 + e^{-z}}$ <br>

- `relu`: $max(0, z)$ <br>
- `linear`: $z$

تغییر  یه بخش از کد خودش

````python
# ---------------------------------------------------------------------
# 1) Activation function for the hidden layer
# ---------------------------------------------------------------------
def activation_fn(Z: np.ndarray, kind: str = "sigmoid") -> np.ndarray:
    """
    Compute element-wise activation over Z.

    TODO (students must implement):
      - Support the following activations: 'sigmoid', 'tanh', 'relu', 'linear'
      - Return a NumPy array of the SAME shape as Z.
      - Validate 'activation' and raise ValueError for unknown activations.

    Args:
      Z   : np.ndarray of shape [N, L] (pre-activation hidden responses)
      kind: str, one of {'sigmoid', 'tanh', 'relu', 'linear'}

    Returns:
      H   : np.ndarray of shape [N, L] (post-activation hidden responses)
    """
    # TODO: implement each activation branch and return the result
    if kind == "sigmoid":
        return 1 / (1 + np.exp(-Z))

    elif kind == "tanh":
        return np.tanh(Z)

    elif kind == "relu":
        return np.maximum(0, Z)

    elif kind == "linear":
        return Z

    else:
        raise ValueError(f"Unknown activation function: '{kind}'. "
                         f"Supported: 'sigmoid', 'tanh', 'relu', 'linear'")
````

**۵) شبه‌وارون یا حل ریج (Ridge / Pseudo–Inverse Solver)**

در مرحله‌ی آموزش ELM، وزن‌های خروجی از رابطه‌ی کمینه‌سازی زیر بدست می‌آیند:

$$
min_{\beta} \|H\beta - Y\|_2^2
$$

که راه‌حل تحلیلی آن $\beta = H^{+}Y$ است؛  
در آن $H^{+}$ همان شبه‌معکوس مور–پنروز (Moore–Penrose Pseudo–Inverse) است.  
اما در عمل، گاهی ماتریس $H^T H$ **تکین (Singular)** یا **بدشرط (Ill‑conditioned)** است، و محاسبهٔ مستقیم شبه‌معکوس ناپایدار می‌شود.  

برای افزایش پایداری عددی، از **رگرسیون ریج (Ridge Regression)** استفاده می‌کنیم که فرم منظم‌شده‌ی مسئله است:

$$
\min_{\beta} \|H\beta - Y\|_2^2 + \lambda \|\beta\|_2^2
$$

حل بستهٔ این رابطه برابر است با:

$$
\beta = (H^{\top}H + \lambda I)^{-1} H^{\top} Y
$$

در این‌جا:
- $\lambda > 0$ یک **ضریب تنظیم (Regularization Coefficient)** است که از ناپایداری جلوگیری می‌کند،
- و با افزایش مقدار $\lambda$، مدلی نرم‌تر و با واریانس کمتر بدست می‌آید.



````python

# ---------------------------------------------------------------------
# 5) Moore–Penrose pseudo-inverse solver (least-squares)
# ---------------------------------------------------------------------
def pinv_solver(H: np.ndarray) -> np.ndarray:
    """
    Compute the Moore–Penrose pseudo-inverse of H.

    TODO (students must implement):
      - Use np.linalg.pinv(H) and return it.

    Args:
      H: np.ndarray of shape [N, L]

    Returns:
      H_pinv: np.ndarray of shape [L, N] (the pseudo-inverse of H)
    """
    # >>> TODO: return np.linalg.pinv(H)
    H_pinv = np.linalg.pinv(H)
    return H_pinv

    raise NotImplementedError("pinv_solver: return np.linalg.pinv(H)")


# ---------------------------------------------------------------------
# 5) Ridge solver
#    Returns the matrix A such that Beta = A @ Y
# ---------------------------------------------------------------------
def ridge_solver(H: np.ndarray, lam: float = 1e-4) -> np.ndarray:
    """
    Build the ridge-regression 'solver' matrix A = (H^T H + lam I)^{-1} H^T.

    TODO (students must implement):
      - Compute G = H^T H    (shape [L, L])
      - Add lam * I to G
      - Invert the result and multiply by H^T
      - Return A of shape [L, N]
    Notes:
      - This returns ONLY the left multiplication matrix; to get Beta do: Beta = A @ Y.
      - Choose lam > 0 for numerical stability.

    Args:
      H  : np.ndarray of shape [N, L]
      lam: float, ridge regularization coefficient

    Returns:
      A  : np.ndarray of shape [L, N]
    """
    # >>> TODO: implement ridge closed-form matrix
    L = H.shape[1]
    HtH = H.T @ H
    HtH_reg = HtH + lam * np.eye(L)
    A = np.linalg.inv(HtH_reg) @ H.T
    return A

    raise NotImplementedError("ridge_solver: build (H^T H + lam I)^{-1} H^T")
````

در یادگیری ماشین، عملکرد مدل را می‌توان از دو دیدگاه بررسی کرد:

- **بایاس (Bias):** میزان خطایی که به‌دلیل ساده‌سازی بیش از حد مدل رخ می‌دهد.  
  (مدل نمی‌تواند روابط واقعی داده را به‌خوبی یاد بگیرد.)
- **واریانس (Variance):** میزان حساسیت مدل نسبت به داده‌های آموزشی.  
  (مدل بیش از حد به داده‌های آموزش وابسته می‌شود و روی داده‌های جدید عملکرد ضعیفی دارد.)

مدلی با **بایاس زیاد** معمولاً ساده است و دچار **کم‌برازش(Underfitting)** می‌شود،  
درحالی‌که مدلی با **واریانس زیاد** معمولاً پیچیده است و **بیش‌برازش (Overfitting)** می‌کند.

- اگر مقدار $\lambda$ خیلی بزرگ باشد، چه اتفاقی برای وزن‌ها و خروجی مدل می‌افتد؟  <br>

**مراحل آموزش**

۱. **مقداردهی اولیهٔ پارامترها:**  
   وزن‌های ورودی و بایاس مطابق تابع `init_params(d, L, seed)` تولید می‌شوند:
   $$
   W \in \mathbb{R}^{d\times L}, \quad b \in \mathbb{R}^{1\times L}
   $$

۲. **محاسبهٔ پاسخ لایه‌ی پنهان:**  
   با استفاده از تابع فعال‌سازی انتخابی، ماتریس پاسخ لایه‌ی پنهان به‌صورت زیر ساخته می‌شود:
   $$
   H = g(XW + b)
   $$
   که در آن $g(\cdot)$ همان تابع فعال‌سازی است.

۳. **حل تحلیلی برای وزن‌های خروجی $\beta$ :**

   برای تخمین وزن‌های نهایی، یکی از دو روش زیر به‌کار می‌رود.

   **الف)** روش شبه‌وارون (Least Squares):
   $$
   \beta = H^{+} Y
   $$

   **ب)** روش ریج (Ridge-Regularized Solution):
   $$
   \beta = (H^{\top}H + \lambda I)^{-1} H^{\top} Y
   $$
   که در آن $\lambda > 0$ عامل تنظیم و پایداری عددی است.

۴. **ساخت مدل خروجی:**  
   مدل با مقادیر زیر بازگردانده می‌شود:
```python
   model = {"W": W, "b": b, "beta": beta, "activation": activation}


````python

# ---------------------------------------------------------------------
# 6) Fit ELM (train output weights Beta analytically)
# ---------------------------------------------------------------------
def fit_elm(
    X: np.ndarray,
    Y: np.ndarray,
    L: int = 20,
    activation: str = "sigmoid",
    use_ridge: bool = False,
    lam: float = 1e-4,
    seed: int | None = 42,
) -> dict:
    """
    Train a minimal ELM:
      1) Initialize W, b
      2) Compute H = g(XW + b)
      3) Solve for Beta:
         - if use_ridge: Beta = (H^T H + lam I)^{-1} H^T Y
         - else        : Beta = pinv(H) @ Y

    TODO (students must implement):
      - Call init_params(d, L, seed) with d = X.shape[1]
      - Build H via hidden_response(X, W, b, activation)
      - Compute Beta using either ridge_solver or pinv_solver
      - Return a model dict containing 'W', 'b', 'beta', and 'activation'

    Args:
      X         : np.ndarray of shape [N, d]
      Y         : np.ndarray of shape [N, m]
      L         : int, number of hidden units
      activation: str, activation name for hidden layer
      use_ridge : bool, if True use ridge closed-form; otherwise use pseudo-inverse
      lam       : float, ridge coefficient (used only if use_ridge=True)
      seed      : optional int, RNG seed

    Returns:
      model     : dict with keys {'W', 'b', 'beta', 'activation'}
    """
    # >>> TODO: tie everything together and return model dict

        # >>> TODO: tie everything together and return model dict
    # 1. گرفتن بعد ورودی (d)
    d = X.shape[1]

    # 2. مقداردهی اولیه پارامترهای ورودی
    W, b = init_params(d, L, seed)

    # 3. محاسبه ماتریس پاسخ لایه پنهان
    H = hidden_response(X, W, b, activation)

    # 4. محاسبه وزن‌های خروجی β
    if use_ridge:
        # روش ریج
        A = ridge_solver(H, lam)
        beta = A @ Y
    else:
        # روش شبه‌وارون
        H_pinv = pinv_solver(H)
        beta = H_pinv @ Y

    # 5. ساخت و بازگرداندن مدل
    model = {
        "W": W,
        "b": b,
        "beta": beta,
        "activation": activation
    }

    return model

    raise NotImplementedError("fit_elm: implement full training pipeline")

````

**۱)تابع فعال‌سازی (Activation Function)**

در این بخش باید توابع فعال‌سازی پایه‌ای را برای استفاده در شبکهٔ MLP پیاده‌سازی کنید.

**وظیفه:**  
سه تابع زیر و مشتق آن‌ها را بنویسید تا در مرحلهٔ پس‌انتشار خطا (Backpropagation) قابل استفاده باشند.


````python

# ---------------------------------------------------------------------
# 1) Activation function and derivative
# ---------------------------------------------------------------------
def activation_fn(Z: np.ndarray, kind: str = "relu") -> np.ndarray:
    """
    Compute the element-wise activation.

    TODO (students must implement):
      - Support 'relu', 'tanh', 'sigmoid'
      - Return same shape as Z.

    Args:
      Z   : np.ndarray [N, units]
      kind: activation type

    Returns:
      A   : np.ndarray [N, units]
    """
    if kind == "relu":
        return np.maximum(0, Z)
    elif kind == "tanh":
        return np.tanh(Z)
    elif kind == "sigmoid":
        return 1 / (1 + np.exp(-Z))
    else:
        raise ValueError(f"Unknown activation function: '{kind}'. "
                         f"Supported: 'relu', 'tanh', 'sigmoid'")


def activation_derivative(A: np.ndarray, kind: str = "relu") -> np.ndarray:
    """
    Compute derivative of activation with respect to pre-activation Z.

    TODO (students must implement):
      - Use A (post-activation values) to compute derivative efficiently.
      - For:
         relu: 1 if A>0 else 0
         tanh: 1 - A^2
         sigmoid: A * (1 - A)

    Args:
      A   : np.ndarray [N, units]
      kind: activation type

    Returns:
      dA  : np.ndarray [N, units]
    """
    if kind == "relu":
        # مشتق ReLU: 1 اگر A>0، در غیر این صورت 0
        return (A > 0).astype(float)
    elif kind == "tanh":
        # مشتق tanh: 1 - A^2
        return 1 - A**2
    elif kind == "sigmoid":
        # مشتق sigmoid: A * (1 - A)
        return A * (1 - A)
    else:
        raise ValueError(f"Unknown activation function: '{kind}'. "
                         f"Supported: 'relu', 'tanh', 'sigmoid'")

````

````python
# ---------------------------------------------------------------------
# 2) Parameter initialization
# ---------------------------------------------------------------------
def initialize_parameters(input_dim: int, hidden_dim: int, output_dim: int, seed: int = 0) -> dict:
    """
    Initialize weights and biases for a 1-hidden-layer MLP.

    TODO (students must implement):
      - Use small random normal values for weights.
      - Biases should start at zero.
      - Shapes:
          W1: [input_dim, hidden_dim]
          b1: [1, hidden_dim]
          W2: [hidden_dim, output_dim]
          b2: [1, output_dim]

    Returns:
      params: dict containing W1, b1, W2, b2
    """
    # TODO: seed (optional), then sample W and b
    np.random.seed(seed)

    # وزن‌های لایه اول
    W1 = np.random.randn(input_dim, hidden_dim) * 0.01
    b1 = np.zeros((1, hidden_dim))

    # وزن‌های لایه دوم
    W2 = np.random.randn(hidden_dim, output_dim) * 0.01
    b2 = np.zeros((1, output_dim))

    params = {
        "W1": W1,
        "b1": b1,
        "W2": W2,
        "b2": b2
    }

    return params

    raise NotImplementedError("initialize_parameters: init small random weights and zeros bias")
````

**۳)انتشار رو به جلو (Forward Pass)**

در این مرحله باید خروجی شبکه را با عبور داده از لایه‌ها محاسبه کنید.  
ابتدا فعال‌سازی لایهٔ پنهان و سپس خروجی نهایی را بسازید.

تمام مقادیر میانی را در دیکشنری `cache` ذخیره کنید تا در مرحلهٔ پس‌انتشار قابل استفاده باشند.


````python

def forward_pass(X: np.ndarray, params: dict, activation: str = "relu") -> dict:
    """
    Compute forward pass through the network:
        Z1 = X @ W1 + b1
        A1 = activation_fn(Z1)
        Z2 = A1 @ W2 + b2
        A2 = softmax or linear (depending on output)

    TODO (students must implement):
      - Compute all intermediate values and store in a cache dict.

    Returns:
      cache: dict containing Z1, A1, Z2, A2
    """
    # استخراج پارامترها
    W1 = params["W1"]
    b1 = params["b1"]
    W2 = params["W2"]
    b2 = params["b2"]

    # لایه پنهان
    Z1 = X @ W1 + b1  # تبدیل خطی
    A1 = activation_fn(Z1, kind=activation)  # فعال‌سازی

    # لایه خروجی
    Z2 = A1 @ W2 + b2  # تبدیل خطی

    # تعیین نوع خروجی (از params یا فرض classification)
    task = params.get("task", "classification")

    if task == "classification":
        # Softmax برای طبقه‌بندی
        # برای پایداری عددی: subtract max
        exp_Z2 = np.exp(Z2 - np.max(Z2, axis=1, keepdims=True))
        A2 = exp_Z2 / np.sum(exp_Z2, axis=1, keepdims=True)
    else:
        # Linear برای رگرسیون
        A2 = Z2

    # ذخیره مقادیر میانی
    cache = {
        "Z1": Z1,
        "A1": A1,
        "Z2": Z2,
        "A2": A2
    }

    return cache

    raise NotImplementedError("forward_pass: implement basic MLP forward propagation")
````

**۴)تابع هزینه (Loss Function)**

در این بخش باید اختلاف بین خروجی پیش‌بینی‌شده و مقدار واقعی را محاسبه کنید.  
تابع هزینه میزان خطای مدل را مشخص می‌کند و مبنای به‌روزرسانی وزن‌هاست.  

دو حالت کلی:  
- **طبقه‌بندی (Classification):** <br>
 از **Cross-Entropy Loss** استفاده کنید  

  $$L = -\frac{1}{N} \sum_i y_i \log(\hat{y_i})$$

- **رگرسیون (Regression):** <br>
 از **Mean Squared Error (MSE)** استفاده کنید  

  $$L = \frac{1}{N} \sum_i (y_i - \hat{y_i})^2$$

</div>

````python

# ---------------------------------------------------------------------
# 4) Loss function
# ---------------------------------------------------------------------
def compute_loss(Y_true: np.ndarray, Y_pred: np.ndarray, task: str = "classification") -> float:
    """
    Compute loss value.

    TODO (students must implement):
      - For classification: use cross-entropy
      - For regression: use MSE

    Args:
      Y_true: [N, m]
      Y_pred: [N, m]
      task: "classification" or "regression"

    Returns:
      loss: float
    """
    N = Y_true.shape[0]  # تعداد نمونه‌ها

    if task == "classification":
        # Cross-entropy loss
        # جلوگیری از log(0) با clipping
        Y_pred = np.clip(Y_pred, 1e-15, 1 - 1e-15)

        # محاسبه cross-entropy: -1/N * Σ(y_true * log(y_pred))
        loss = -np.sum(Y_true * np.log(Y_pred)) / N

    elif task == "regression":
        # Mean Squared Error (MSE)
        # 1/N * Σ(y_true - y_pred)^2
        loss = np.mean((Y_true - Y_pred) ** 2)

    else:
        raise ValueError(f"Unknown task: '{task}'. "
                         f"Supported: 'classification', 'regression'")

    return loss

    raise NotImplementedError("compute_loss: implement cross-entropy and MSE")
````

ابتدا گرادیان‌های لایه‌ی خروجی، سپس لایه‌ی پنهان را به دست آورید و در دیکشنری `grads` ذخیره کنید.  
دقت کنید که ابعاد تمام ماتریس‌ها در هر مرحله با داده‌ها و پارامترها سازگار باشد.


````python

def backward_pass(X: np.ndarray, Y: np.ndarray, params: dict, cache: dict,
                  activation: str = "relu", task: str = "classification") -> dict:
    """
    Compute gradients for all parameters using backpropagation.

    TODO (students must implement):
      - Derive dZ2, dW2, db2
      - Derive dA1, dZ1, dW1, db1
      - Use chain rule carefully.
      - Check shapes after each step.

    Returns:
      grads: dict containing dW1, db1, dW2, db2
    """
    # استخراج پارامترها و مقادیر cache
    W1, b1, W2, b2 = params["W1"], params["b1"], params["W2"], params["b2"]
    Z1, A1, Z2, A2 = cache["Z1"], cache["A1"], cache["Z2"], cache["A2"]

    N = X.shape[0]  # تعداد نمونه‌ها

    # ========== گرادیان‌های لایه خروجی ==========

    if task == "classification":
        # برای Cross-Entropy + Softmax: dZ2 = A2 - Y
        dZ2 = A2 - Y  # شکل: (N, output_dim)
    else:
        # برای MSE: dZ2 = 2 * (A2 - Y) / N
        dZ2 = 2 * (A2 - Y) / N  # شکل: (N, output_dim)

    # گرادیان وزن‌های لایه خروجی
    dW2 = (A1.T @ dZ2) / N  # شکل: (hidden_dim, output_dim)

    # گرادیان بایاس لایه خروجی
    db2 = np.sum(dZ2, axis=0, keepdims=True) / N  # شکل: (1, output_dim)

    # ========== گرادیان‌های لایه پنهان ==========

    # گرادیان نسبت به خروجی لایه پنهان
    dA1 = dZ2 @ W2.T  # شکل: (N, hidden_dim)

    # گرادیان نسبت به ورودی لایه پنهان (قبل از فعال‌سازی)
    dZ1 = dA1 * activation_derivative(A1, kind=activation)  # شکل: (N, hidden_dim)

    # گرادیان وزن‌های لایه پنهان
    dW1 = (X.T @ dZ1) / N  # شکل: (input_dim, hidden_dim)

    # گرادیان بایاس لایه پنهان
    db1 = np.sum(dZ1, axis=0, keepdims=True) / N  # شکل: (1, hidden_dim)

    # ذخیره گرادیان‌ها
    grads = {
        "dW1": dW1,
        "db1": db1,
        "dW2": dW2,
        "db2": db2
    }

    return grads

    raise NotImplementedError("backward_pass: compute gradients for all params")

````

**۶) به‌روزرسانی پارامترها**

در این مرحله وزن‌ها و بایاس‌ها را با قانون گرادیان نزولی(Gradient Descent) به‌روزرسانی کنید:

$$
\theta \leftarrow \theta - \eta\, \frac{\partial L}{\partial \theta}
$$

برای هر پارامتر از جمله $W_1, b_1, W_2, b_2$ همین فرمول اعمال می‌شود تا مقدار آن اصلاح گردد.


````python

# ---------------------------------------------------------------------
# 6) Parameter update (Gradient Descent)
# ---------------------------------------------------------------------
def update_parameters(params: dict, grads: dict, lr: float = 0.01) -> dict:
    """
    Update parameters using gradient descent.

    TODO (students must implement):
      - params = params - lr * grads
      - Update all W and b.

    Returns:
      params: updated parameter dict
    """
    # به‌روزرسانی وزن‌ها و بایاس‌های لایه پنهان
    params["W1"] -= lr * grads["dW1"]
    params["b1"] -= lr * grads["db1"]

    # به‌روزرسانی وزن‌ها و بایاس‌های لایه خروجی
    params["W2"] -= lr * grads["dW2"]
    params["b2"] -= lr * grads["db2"]

    return params

    raise NotImplementedError("update_parameters: implement gradient descent update")

````

**۷)حلقه‌ی آموزش شبکه**

در این بخش باید کل فرآیند آموزش MLP را پیاده‌سازی کنید.  
این فرآیند شامل چند مرحله‌ی تکرارشونده (epoch) است که در هر مرحله اتفاقات زیر رخ می‌دهد:  

۱. مقداردهی اولیه‌ی پارامترها  
۲. اجرای انتشار رو‌به‌جلو  
۳. محاسبه‌ی خطا (Loss)  
۴. اجرای پس‌انتشار خطا  
۵. به‌روزرسانی پارامترها  

در صورت تمایل می‌توانید مقدار خطا را هر چند epoch چاپ کنید تا روند یادگیری مدل مشخص شود.  
هدف این است که با افزایش epoch، مقدار $Loss$ کاهش یابد.

</div>

````python

# ---------------------------------------------------------------------
# 7) Training loop
# ---------------------------------------------------------------------
def train_mlp(
    X: np.ndarray,
    Y: np.ndarray,
    hidden_dim: int,
    activation: str = "relu",
    task: str = "classification",
    epochs: int = 1000,
    lr: float = 0.01,
    seed: int = 0,
    verbose: bool = True
) -> dict:
    """
    Full training loop for a 1-hidden-layer MLP.

    TODO (students must implement):
      - Initialize parameters
      - For each epoch:
         - Forward pass
         - Compute loss
         - Backward pass
         - Update parameters
      - (Optional) print loss every few epochs

    Returns:
      params: trained model parameters
    """
    # تعیین ابعاد ورودی و خروجی
    input_dim = X.shape[1]
    output_dim = Y.shape[1]

    # ۱. مقداردهی اولیه پارامترها
    params = initialize_parameters(input_dim, hidden_dim, output_dim, seed)
    params["task"] = task  # ذخیره نوع task در params

    # لیست برای ذخیره loss هر epoch (اختیاری)
    losses = []

    # ۲. حلقه آموزش
    for epoch in range(epochs):
        # الف) Forward pass
        cache = forward_pass(X, params, activation)

        # ب) محاسبه loss
        loss = compute_loss(Y, cache["A2"], task)
        losses.append(loss)

        # ج) Backward pass
        grads = backward_pass(X, Y, params, cache, activation, task)

        # د) به‌روزرسانی پارامترها
        params = update_parameters(params, grads, lr)

        # نمایش loss هر ۱۰۰ epoch (اختیاری)
        if verbose and (epoch % 100 == 0 or epoch == epochs - 1):
            print(f"Epoch {epoch:4d}/{epochs} - Loss: {loss:.6f}")

    # نمایش loss نهایی
    if verbose:
        print(f"\nآموزش کامل شد!")
        print(f"Loss اولیه: {losses[0]:.6f}")
        print(f"Loss نهایی:  {losses[-1]:.6f}")
        print(f"تعداد کاهش:  {losses[0] - losses[-1]:.6f}")

    return params

    raise NotImplementedError("train_mlp: implement simple training loop")

````

**۸) پیش‌بینی با مدل آموزش‌دیده**

در این مرحله باید از مدل آموزش‌دیده برای پیش‌بینی بر روی داده‌های جدید استفاده کنید.  
ابتدا انتشار رو‌به‌جلو را با پارامترهای نهایی انجام دهید تا خروجی شبکه به‌دست آید.  

سپس:  
- اگر مسئله **طبقه‌بندی** است: با استفاده از `argmax` بر روی خروجی، کلاس نهایی را انتخاب کنید.  
- اگر مسئله **رگرسیون** است: مقدار خام خروجی مدل را برگردانید.

````python

# ---------------------------------------------------------------------
# 8) Prediction
# ---------------------------------------------------------------------
def predict_mlp(X: np.ndarray, params: dict, activation: str = "relu", task: str = "classification") -> np.ndarray:
    """
    Predict outputs from trained MLP.

    TODO (students must implement):
      - Forward pass
      - If classification: return argmax of probabilities
      - If regression: return raw predictions

    Returns:
      Y_pred: predicted labels or values
    """
    # انجام forward pass با پارامترهای آموزش‌دیده
    cache = forward_pass(X, params, activation)

    # تعیین task (از params بگیر یا از ورودی تابع)
    if "task" in params:
        current_task = params["task"]
    else:
        current_task = task

    # بر اساس نوع مسئله، خروجی مناسب را برگردان
    if current_task == "classification":
        # برای طبقه‌بندی: اندیس کلاس با بیشترین احتمال
        y_pred = np.argmax(cache["A2"], axis=1)
        return y_pred
    elif current_task == "regression":
        # برای رگرسیون: مقادیر پیش‌بینی شده
        return cache["A2"]
    else:
        raise ValueError(f"Unknown task: '{current_task}'. "
                         f"Supported: 'classification', 'regression'")
    raise NotImplementedError("predict_mlp: implement inference logic")
````
